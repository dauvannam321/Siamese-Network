**Import library**

```python
import os
import glob
```

```python
import torch
import torchvision.transforms as transforms
from torchvision.utils import save_image
from PIL import Image
import random
import shutil
import zipfile
from google.colab import files
```

```python
root_data_path = '/content/Half_oxbuild_images'
```

**Augment methods**

```python
transforms_list = [
    transforms.RandomRotation(degrees=15),
    transforms.RandomHorizontalFlip(p=0.5)
]
```

```python
# Create a random apply transform
random_transform = transforms.RandomApply(transforms_list, p=0.5)
```

```python
for sub_folder in os.listdir(root_data_path):
  print(sub_folder) # oriel
  folder_path = os.path.join(root_data_path, sub_folder)
  # print(folder_path) # /content/Half_oxbuild_images/oriel
  file_extension = '*.jpg' # Replace with the file extension of your images

  file_list = glob.glob(os.path.join(folder_path, file_extension)) # ['/content/Half_oxbuild_images/oriel/oriel_000006.jpg', ... ]
  # image_path = file_list[0]
  # print(image_path) # /content/Half_oxbuild_images/oriel/oriel_000006.jpg
  # print(image_path.split('/')[4]) # oriel_000006.jpg

  new_path_to_save_img = '/content/' + sub_folder
  if not os.path.exists(new_path_to_save_img):
    os.mkdir(new_path_to_save_img)
  for i in range(701 - len(file_list)):
    image_path = random.choice(file_list)
    image = Image.open(image_path)
    transformed_image = random_transform(image)
    transformed_image.save(os.path.join(new_path_to_save_img,'augment_' + str(i) + '_' + image_path.split('/')[4]))
```

```python
for sub_folder in os.listdir(root_data_path):
  print(sub_folder) # oriel
  folder_path = os.path.join(root_data_path, sub_folder)
  # print(folder_path) # /content/Half_oxbuild_images/oriel
  file_extension = '*.jpg' # Replace with the file extension of your images

  new_path = '/content/' + sub_folder
  file_list = glob.glob(os.path.join(new_path, file_extension)) # ['/content/Half_oxbuild_images/oriel/oriel_000006.jpg', ... ]
  # print(file_list)
  # print(file_list[0].split('/')[3])
  for img_path in file_list:
    destination_file = folder_path + '/' + img_path.split('/')[3]
    shutil.copyfile(img_path, destination_file)
print("Done!")
```

```python
# specify the path to the folder to be downloaded
folder_path = "/content/Half_oxbuild_images"

# create a zipfile object to compress the folder
zip_name = "700_oxbuild.zip"
zip_path = os.path.join("/content", zip_name)
zip_file = zipfile.ZipFile(zip_path, "w", zipfile.ZIP_DEFLATED)

# iterate through each file in the folder and add it to the zipfile
for foldername, subfolders, filenames in os.walk(folder_path):
  for filename in filenames:
    file_path = os.path.join(foldername, filename)
    zip_file.write(file_path)

# close the zipfile object
zip_file.close()

# download the zipfile using the files module
files.download(zip_path)
```
